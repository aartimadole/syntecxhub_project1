```python
import pandas as pd
```

### 1. Read CSV / Excel into DataFrame


```python
df = pd.read_csv(r"E:\Syntecxhub_task1_pandas\Sample - Superstore.csv", encoding="latin1")

print("First 5 rows:\n", df.head())
print("\nLast 5 rows:\n", df.tail())
print("\nData Types:\n", df.dtypes)
```

### 2. Compute summary statistics


```python
summary_stats = df.describe()
print("\nSummary Statistics:\n", summary_stats)
```

```python
print("üîπ Mean Sales:", df["Sales"].mean())
print("üîπMedian sales:", df["Sales"].median())
print("üîπMinimum Sales:",df["Sales"].min())
print("üîπMaximum Sales:", df["Sales"].max())
print("üîπTotal Orders:",df["Order ID"].count())
      
```

### 3.Filtering Example


```python
# High sales orders (Sales > 1000)
high_sales = df[df["Sales"] > 1000]

print("\nüîπ High Sales Orders (Sales > 1000):")
print("high_sales.head()")

```

### 5.Select Specific Columns



```python


sales_profit = df[["Order ID", "Customer Name", "Sales", "Profit"]]

print("\nüîπ Selected Columns:")
print(sales_profit.head())
```

### 6Ô∏è‚É£ Slice Subset (First 10 Rows)


```python
subset = df.iloc[0:10]

print("\nüîπ First 10 Rows:")
print(subset)
```

###  7Ô∏è‚É£ GroupBy Analysis (Business Insight)


```python

region_sales = df.groupby("Region")["Sales"].sum()

print("\nüîπ Total Sales by Region:")
print(region_sales)

category_profit = df.groupby("Category")["Profit"].sum()

print("\nüîπ Total Profit by Category:")
print(category_profit)
```

### 8Ô∏è‚É£ Save Filtered Results


```python

high_sales.to_csv("high_sales_orders.csv", index=False)
region_sales.to_csv("region_sales_summary.csv")

print("\n‚úÖ Analysis Completed & Files Saved Successfully!")
```
